# Prática 01 - Frutas

```python
# Bibliotecas

import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Carregando os dados
df = pd.read_excel("../data/dados_frutas.xlsx")
df.head(10)
```

```python
# Variáveis independentes e dependentes 
y = df["Fruta"] # Target
X = df.drop("Fruta", axis = 1) # Features

display(X)
display(y)
```

```python
# Vamos ensinar a máquina a aprender a classificar as frutas, para isso, vamos usar o algoritmo de árvore de decisão.
arvore = tree.DecisionTreeClassifier() # Criando o objeto da árvore de decisão

# Aqui que temos o Machine Learning (o problema é chegar aqui)
arvore.fit(X, y) # Treinando a árvore de decisão com os dados de treino (X e y)
```

```python
# Com base no treinamento, vamos predizer uma fruta
resultado = arvore.predict([[0, 1, 1, 0]])
print(resultado)
```

```python
# Vamos desenhar a árvore de decisão
plt.figure(dpi = 150) # Aumentando a resolução da figura

tree.plot_tree(
    arvore, # Modelo de árvore já treinado
    feature_names = X.columns, # Nomes das características usadas nas decisões
    class_names = arvore.classes_, # Nomes das possíveis classes previstas
    filled = True # Colore os nós conforme a classe predominante e pureza
)

# Perceba que os nós folhas são pintados pelo filled = True, e a cor é definida pela classe predominante e pureza do nó. Quanto mais puro o nó, mais escura a cor. Quanto mais misturado o nó, mais clara a cor. O nó raiz é o nó mais escuro, pois é o nó mais puro, pois tem todos os dados. Conforme vamos descendo na árvore, os nós vão ficando mais claros, pois vão ficando mais misturados.
```

Para interpretar esse resultado, sempre olhe a pergunta (primeira coisa que aparece no nó). Por exemplo, Suculenta <= 0.5, nesse caso a pergunta é `Não é suculenta?`, uma vez que para satisfazer a inequação, tem que ser não suculenta (True), e como os valores são 0 ou 1, se for 0 é verdadeiro de não ser suculentan então vai para a esquerda.

Além disso, o **Gini** mede o quanto os dados de um nó estão misturados entre diferentes classes. 

Resumo: 
- alguma_coisa <= 0.5: pergunta usada para separar os dados naquele ponto.
- gini = 0.5: ainda há mistura de classes naquele nó. Quanto mais próximo de 0, mais puro ele é.
- samples = 2: duas frutas chegaram até aquele nó.
- value = [...]: quantidade de exemplos de cada classe naquele nó, seguindo a ordem de arvore.classes_.
- class = Banana: classe que a árvore preveria caso parasse naquele nó. Como há empate nesse ponto, aparece uma das classes segundo a ordem interna das classes.

```python
# Ordem das classes que são representadas no `value`:
print(arvore.classes_) # Ordem alfabética por padrão, mas pode ser alterada usando o parâmetro class_names na função plot_tree.
```

```python
# Usar o predict com probabilidade

# Como alternativa à saída de uma classe específica, a probabilidade de cada classe pode ser previsto, que é a fração de amostras de treinamento da classe em um folha:
resultado = arvore.predict_proba([[0, 1, 1, 1]])
print(resultado)
```

O resultado quando colocamos que não é arredondada, mas é suculenta	vermelh e doce, dá com 100% de certeza que é morango.

```python
resultado = arvore.predict_proba([[1, 1, 1, 1]])[0] # Para pegar o array mais interno (veja o resultado acima, e perceba que tem dois [[]])
pd.Series(resultado, index = arvore.classes_)
```

Não existe uma característica diferente para separar o que é uma maça do que é uma cereja, então o algoritmo entende que pode ser uma das duas, mas na árvore, por ordem alfabética, prevalece a cereja.
