# Prática 5 - Classificação

```python
# Bibliotecas
import pandas as pd
from sklearn import linear_model
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Carregando os dados
df = pd.read_excel('../data/dados_cerveja_nota.xlsx', index_col = 0)
df.head()
```

```python
# Vamos criar uma feature nova 
df['aprovado'] = (df['nota'] > 5).astype(int)
df.head()
```

```python
# Vamos plotar o gráfico disso
plt.plot(df['cerveja'], df['aprovado'], 'o')
plt.grid()
plt.title('Cerveja vs Aprovação')
plt.xlabel('Cervejas')
plt.ylabel('Aprovado')
```

## Regressão logística

```python
# Criando o modelo
regressao = linear_model.LogisticRegression(
    penalty = None,
    fit_intercept = True 
)

# Treinando o modelo
regressao.fit(X = df[['cerveja']], y = df['aprovado'])
```
